### Broadcasting in NumPy

The term **<u>broadcasting</u>** describes how numpy treats arrays with different shapes during <u>arithmetic operations</u>. Subject to certain constraints, the smaller array is “broadcast” across the larger array so that they have compatible shapes. Broadcasting provides a means of vectorizing array operations.

#### General rules for Broadcasting
When operating on two arrays, NumPy compares their shapes element-wise. It starts with the trailing dimensions and works its way forward. Two dimensions of size N are compatible when <br>

they are equal, or <br>
one of them is the singular value <br>
one is N * N and other is N * 1 <br>
one is N * N and other is 1 * N 

### When one operand is N*N and other is 1 * 1

```python
import numpy as np
```

```python
Z1 = np.arange(9).reshape(3,3)
Z2 = 1 
print(Z1) 

print(Z1 + Z2)
```

### When one operand is N * N and other is N * 1

```python
Z1 = np.arange(9).reshape(3,3)
Z2 = np.arange(3)[::-1].reshape(3,1)
# print(Z1 + Z2)
Z2
```
